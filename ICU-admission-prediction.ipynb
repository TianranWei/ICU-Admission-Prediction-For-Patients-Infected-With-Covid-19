```python
import pandas as pd
import numpy as np
import warnings
import googletrans
from googletrans import Translator

pd.set_option('display.max_columns', None) # shows all columns
warnings.filterwarnings('ignore')
```

```python
# Function 1
# Since the columns are still in Spanish, write a function to translate it to English
def translate(df):
    translator = Translator()
    df.rename(columns=lambda x: translator.translate(x).text, inplace=True)
    return df
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/marianarf/covid19_mexico_analysis/master/mexico_covid19.csv')
data = translate(data)
```
